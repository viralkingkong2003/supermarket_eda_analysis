```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
db_url=os.getenv("DATABASE_URL")
```

```python
load_dotenv()
con =os.getenv("DATABASE_URL")
```

```python
engine = create_engine(con)
```

```python
tquery= '''
select * from
 public.transaction_data;
 '''  
transaction_data = pd.read_sql(tquery,con=engine)
```

```python
pquery= '''
 select* from  
	public.super_data sd;
    '''

purchase_data = pd.read_sql(pquery,con=engine)
```

OBJECTIVES 
1. top 10 most purchased brands
2. total sales by customer life stage

```python
supermarket_data=purchase_data.merge(transaction_data,on='LYLTY_CARD_NBR',how='right')
```

```python
top_brands = (supermarket_data.groupby('PROD_NAME')['PROD_QTY'].sum().sort_values(ascending=False).head(10))
```

Monthly sales trend

```python
supermarket_data['YEAR']=supermarket_data['DATE'].dt.year
```

```python
supermarket_data['MONTH']=supermarket_data['DATE'].dt.month
```

```python
supermarket_data['MONTH_NAME']=supermarket_data['DATE'].dt.month_name()
```

```python
supermarket_data.head()
```

```python
monthly_sales=(supermarket_data.groupby(['YEAR','MONTH_NAME'])['TOT_SALES'].sum().sort_values(ascending=False))
```

```python
monthly_sales=(supermarket_data.groupby(['YEAR','MONTH_NAME'])['TOT_SALES'].sum().reset_index().sort_values(['YEAR','MONTH_NAME']))
```

```python
monthly_sales
```

```python
yearly_sales=(supermarket_data.groupby(['YEAR'])['TOT_SALES'].sum().sort_values(ascending=False))
```

```python
yearly_sales
```
